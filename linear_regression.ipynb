```
import numpy as np
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
```

```
x, y, theta_true = make_regression(n_samples=100,
                                   n_features=1,
                                   bias=1,
                                   noise=10,
                                   coef=True,
                                   shuffle=True,
                                   random_state=123)
plt.scatter(x, y)
```

```
theta_hat = (x.T @ x) ** -1 @ x.T @ y

print(f'True theta: {theta_true}')
print(f'Estimated theta: {theta_hat[0]}')

plt.scatter(x, x * theta_hat)
```
